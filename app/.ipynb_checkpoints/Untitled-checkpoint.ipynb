```python
json_data = json.loads(json.dumps(data['response']['body'], ensure_ascii=False))
```

```python
if type(json_data['items']['item']) == type(list()):
    df = pd.DataFrame(json_data['items']['item'])
elif type(json_data['items']['item']) == type(dict()):
    df = pd.DataFrame([json_data['items']['item']])
df['mainPurpsCdNm']
```
